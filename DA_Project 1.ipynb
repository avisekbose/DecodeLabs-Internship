```python
# Detecting Outliers (Boxplot)

plt.figure()
df.boxplot(column=['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice'])
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
# Removing Outliers
def remove_outliers_iqr(df, columns):
    for column in columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        df = df[(df[column] <= upper)]
    
    return df

df = remove_outliers_iqr(df, ['TotalPrice'])
```

```python
plt.figure()
df.boxplot(column=['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice'])
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
# Ploting an Histogram for data distribution

numerical_cols = ['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice']

# Define grid size (rows, cols)
rows = 2
cols = 2

plt.figure(figsize=(15, 10))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(rows, cols, i)
    plt.hist(df[col], bins=10)
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel("Frequency")

plt.suptitle('Histogram for Data Distribution', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# Identifing correlation

df_corr = df.corr(numeric_only=True)
print(df_corr)

# Ploting a Heatmap

sns.heatmap(df_corr, annot=True)
plt.show()
```
